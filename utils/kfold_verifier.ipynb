```python
import json
import numpy as np
import pandas as pd
import os
from collections import Counter
```

```python
n_split = 4
filter = 0
base_path = f"../dataset/kfold/coco/filter_{filter}/nsplit{n_split}"
categories = {0: "General trash", 1: "Paper", 2: "Paper pack", 3: "Metal", 4: "Glass", 5: "Plastic", 6: "Styrofoam", 7: "Plastic bag", 8: "Battery", 9: "Clothing"}


def get_distribution(y_vals):
        y_distr = Counter(y_vals)
        y_vals_sum = sum(y_distr.values())
        return [f'{y_distr[i] / y_vals_sum:.2%}' for i in range(np.max(y_vals) + 1)]
```

```python
dists = []
index = []

for i in range(n_split):
    train_filename = f"train_cv_{i + 1}"
    val_filename = f"val_cv_{i + 1}"

    with open(os.path.join(base_path, train_filename+".json"), "r") as f:
        train_data = json.load(f)
    with open(os.path.join(base_path, val_filename+".json"), "r") as f:
        val_data = json.load(f)
    train_df = pd.DataFrame.from_dict(train_data["annotations"])
    dists.append(get_distribution(train_df["category_id"]))
    index.append(train_filename)
    dists.append(Counter(train_df["category_id"]).values())
    index.append("")
    val_df = pd.DataFrame.from_dict(val_data["annotations"])
    dists.append(get_distribution(val_df["category_id"]))
    index.append(val_filename)
    dists.append(Counter(val_df["category_id"]).values())
    index.append("")
```

```python
print(f"Area Size Filtered Below {filter} / n_split: {n_split}")
pd.DataFrame(dists, index=index, columns=categories.values())
```
